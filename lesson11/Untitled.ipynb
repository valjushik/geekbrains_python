```python
import sympy
```

```python
x = sympy.Symbol('x')
```

```python
f_x = -12*x**4*sympy.sin(sympy.cos(x)) - 18*x**3 + 5*x**2 + 10*x - 30
```

```python
print(f_x)
```

```python
print(f_x.subs(x, 10.0))
```

```python
sympy.plotting.plot(f_x)
```

```python
diff_f_x = sympy.diff(f_x, x)
```

```python
print(diff_f_x)
```

```python
sympy.plotting.plot(f_x, diff_f_x)
```
